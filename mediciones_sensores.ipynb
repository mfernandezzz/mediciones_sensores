Realizar una limpieza de datos lo mas acertada posible de una hoja de calculo que contiene registros correspondientes a mediciones de sensores.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#cargar la hoja de calculo
df = pd.read_excel('/content/mediciones_sensores.xlsx')
df
```

```python
#pasar la hoja de calculo a un archivo csv
df.to_csv('mediciones_sensores.csv') #index=False, encoding='utf-8'
```

```python
df.head()
```

```python
#visualizar el tipo de dato que almacena cada columna
df.dtypes
```

```python
#comprobar si un id se repite mas de una vez (filas duplicadas)
df['id_medición'].value_counts()
```

```python
#establecer la columna id_medicion como el indice del dataframe
df.set_index('id_medición', inplace=True)
df
```

```python
#comprobar la cantidad de sensores
df['id_sensor'].value_counts()
```

```python
#pasar los id de sensores en minuscula a mayuscula
df['id_sensor'] = df['id_sensor'].str.upper()
df['id_sensor'].value_counts()
```

```python
#buscar errores de tipeo en la columna rayos_uv
df['rayos_uv'].value_counts()
```

```python
#corregir los errores de tipeo en la columna rayos_uv
df['rayos_uv'] = df['rayos_uv'].str.upper()
df['rayos_uv'].value_counts()
```

```python
#visualizar la cantidad de filas que tiene el dataframe
df.shape[0]
```

```python
#eliminar filas duplicadas y ordenar las filas en base a la columna fecha y hora
df = df.drop_duplicates().sort_values('fecha_hora')
df.shape[0]
```

```python
#observar los registros en la columna vel_viento
df['vel_viento'].value_counts()
```

```python
#comprobar si solo se encuentra el string viento moderado en la columna vel_viento
velocidades_viento = df[df['vel_viento'].str.startswith('viento', na=False)]
velocidades_viento.shape[0]
```

```python
#cambiar todos los datos de la columna vel_viento a viento moderado (no parece importante btw)
df['vel_viento'] = 'viento moderado'
df['vel_viento'].value_counts()
```

```python
df.head(10)
```

```python
#cambiar el nombre de la columna temperatura
cols = df.columns.tolist()
cols[cols.index('temperatura')] = 'temperatura (°C)'
df.columns = cols
df.head(10)
```

```python
#verificar si la columna presion contiene valores nulos (nan)
presion_null = np.isnan(df['presión'])
presion_null.value_counts()
```

```python
#verificar si la columna humedad contiene valores nulos
humedad_null = np.isnan(df['humedad'])
humedad_null.value_counts()
```

```python
#obtener un numpy array con todos los valores dentro de la columna humedad
humedad_values = np.array(df['humedad'])
#obtener el valor medio de humedad excluyendo los valores nan
mean_humedad = humedad_values[~np.isnan(humedad_values)].mean()
round(mean_humedad, 1)
```

```python
#obtener el valor minimo de humedad
min_humedad = humedad_values[~np.isnan(humedad_values)].min()
min_humedad
```

```python
#obtener el valor maximo de humedad
max_humedad = humedad_values[~np.isnan(humedad_values)].max()
max_humedad
```

```python
#obtener la desviacion estandar de los valores de humedad
std_humedad = humedad_values[~np.isnan(humedad_values)].std()
round(std_humedad, 1)
```

```python
#sustituir los valores nan por el valor medio de humedad obtenido previamente
df['humedad'] = df['humedad'].fillna(round(mean_humedad, 1))
df.head(10)
```

```python
#verificar si la columna humedad sigue teniendo valores nan
humedad_null = np.isnan(df['humedad'])
humedad_null.value_counts()
```

```python
#obtener el valor medio de humedad con los datos limpiados
round(df['humedad'].mean(), 1)
```

```python
#obtener la desviacion estandar de los valores de humedad con los datos limpiados
round(df['humedad'].std(), 1)
```

```python
#eliminar todos los datos que incluyan ºC en la columna temperatura
df['temperatura (°C)'] = df['temperatura (°C)'].str.replace('°C', '')
df.head(10)
```

```python
#visualizar la cantidad de valores de temperatura registrados en la columna
df['temperatura (°C)'].value_counts()
```

```python
#cambiar los valores de la columna temperatura a type float
df['temperatura (°C)'] = df['temperatura (°C)'].astype(float)
df.dtypes
```

```python
#obtener los valores de temperatura
values_temperatura = np.array(df['temperatura (°C)'])
values_temperatura
```

```python
#obtener el valor medio de temperatura excluyendo los valores nan
mean_temperatura = values_temperatura[~np.isnan(values_temperatura)].mean()
round(mean_temperatura, 1)
```

```python
#obtener el valor minimo de temperatura excluyendo los valores nan
min_temperatura = values_temperatura[~np.isnan(values_temperatura)].min()
min_temperatura
```

```python
#obtener el valor maximo de temperatura excluyendo los valores nan
max_temperatura = values_temperatura[~np.isnan(values_temperatura)].max()
max_temperatura
```

```python
#obtener la desviacion estandar de los valores de temperatura excluyendo los valores nan
std_temperatura = values_temperatura[~np.isnan(values_temperatura)].std()
round(std_temperatura, 1)
```

```python
#sustituir los valores nan por los valores medios de temperatura
df['temperatura (°C)'] = df['temperatura (°C)'].fillna(round(mean_temperatura, 1))
df.head(10)
```

```python
#obtener el valor medio de temperatura con los datos limpiados
round(df['temperatura (°C)'].mean(), 1)
```

```python
#obtener el desvio estandar de temperatura con los datos limpiados
round(df['temperatura (°C)'].std(), 1)
```
