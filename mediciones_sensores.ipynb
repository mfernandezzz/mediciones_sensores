**Verificar si la columna humedad contiene valores nulos**

```python
humedad_null = np.isnan(df['humedad'])
humedad_null.value_counts()
```

Dada la existencia de valores nulos dentro de la columna 'humedad', se procede a obtener el valor medio de los valores validos ingresados dentro de la columna y sustituir los valores nan por el valor medio obtenido.

```python
#obtener un numpy array con todos los valores dentro de la columna humedad
humedad_values = np.array(df['humedad'])
#obtener el valor medio de humedad excluyendo los valores nan
mean_humedad = humedad_values[~np.isnan(humedad_values)].mean()
round(mean_humedad, 1)
```

```python
#obtener el valor minimo de humedad
min_humedad = humedad_values[~np.isnan(humedad_values)].min()
min_humedad
```

```python
#obtener el valor maximo de humedad
max_humedad = humedad_values[~np.isnan(humedad_values)].max()
max_humedad
```

```python
#obtener la desviacion estandar de los valores de humedad
std_humedad = humedad_values[~np.isnan(humedad_values)].std()
round(std_humedad, 1)
```

```python
#sustituir los valores nan por el valor medio de humedad obtenido previamente
df['humedad'] = df['humedad'].fillna(round(mean_humedad, 1))
df.head(10)
```

```python
#verificar si la columna humedad sigue teniendo valores nan
humedad_null = np.isnan(df['humedad'])
humedad_null.value_counts()
```

```python
#obtener el valor medio de humedad con los datos limpiados
round(df['humedad'].mean(), 1)
```

Si se realiza un vistazo rapido a algunos de los datos del dataframe, se puede apreciar que determinada informacion almacenada en las columnas no coincide con la informacion almacenada en otra columna. Por ejemplo, la medicion con id f8f58403 tiene un valor de rayos uv ALTO, lo cual es irracional teniendo en cuenta que la medicion fue realizada a las 04:06. Por consiguiente, se debe llevar adelante una normalizacion de los datos. Para la misma, teniendo en cuenta que las mediciones fueron realizadas durante el verano, se determinara que durante el horario de la noche (a partir de las 21:00) y hasta la salida del sol (aproximadamente las 06:00) el indice de rayos uv sera nulo.

```python
#asignar el valor NULO en la columna rayos uv si en la columna fecha_hora la hora se encuentra en el rango de las 21:00 y 06:00

#especificando valor de hora pero no minutos
#hour = df['fecha_hora'].dt.hour
#cond_noche = (hour >= 21) | (hour <= 6)
#df.loc[cond_noche, 'rayos_uv'] = 'BAJO'

#especificando un rango horario con horas exactas
time = df['fecha_hora'].dt.time
from datetime import time as dtime
start, end = dtime(21, 0, 0), dtime(6, 0, 0)
cond_noche = df['fecha_hora'].dt.time >= start
cond_early = df['fecha_hora'].dt.time <= end
horario_nocturno = cond_noche | cond_early
df.loc[horario_nocturno, 'rayos_uv'] = 'NULO'
df.head(15)
```

```python
#obtener la cantidad de veces que se registro cada nivel de rayos_uv
df['rayos_uv'].value_counts()
```

Si se realiza un nuevo vistazo rapido a los datos, se puede observar que siguen existiendo registros incongruentes entre columnas. Por ejemplo, en el registro con id 8e576280 se registro un indice de rayos uv ALTO, lo cual es irracional teniendo en cuenta que el sensor realizo ese registro a las 06:18 de la mañana. Por consiguiente, no es logico que poco despues del amanecer se registre nivel de rayos uv ALTO. Para esta normalizacion de los datos, se establecera un nivel de radiacion uv BAJO en todas aquellas mediciones relizadas en el horario de 06:01 a 08:00.

```python
start, end = dtime(6, 1, 0), dtime(8, 0, 0)
cond_first_hour = df['fecha_hora'].dt.time >= start
cond_early_morning = df['fecha_hora'].dt.time <= end
horario_mañana = cond_first_hour & cond_early_morning
df.loc[horario_mañana, 'rayos_uv'] = 'BAJO'
df.head(15)
```
