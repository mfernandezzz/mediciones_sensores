Dependiendo del caso, se pueden sustituir los valores nan por el valor de la mediana (50%) o el valor promedio (mean). En este caso puntual, se sustituye por el valor promedio.

```python
values_temperatura = np.array(df['temperatura (°C)'])

#obtener el valor medio de temperatura excluyendo los valores nan
mean_temperatura = values_temperatura[~np.isnan(values_temperatura)].mean()
round(mean_temperatura, 1)
```

Si se realiza un vistazo rapido al dataframe, se puede apreciar que determinados datos almacenados en las columnas no coincide con los datos almacenados en otra columna. Por ejemplo, la medicion con id f8f58403 tiene un valor de rayos uv ALTO, lo cual es irracional teniendo en cuenta que la medicion fue realizada a las 04:06. Por consiguiente, se debe llevar adelante una normalizacion de los datos. Para la misma, teniendo en cuenta que las mediciones fueron realizadas durante el verano, se determinara que durante el horario de la noche (a partir de las 21:00) y hasta la salida del sol (aproximadamente las 06:00) el indice de rayos uv sera NULO.

```python
# asignar el valor NULO en la columna rayos uv si en la columna fecha_hora la hora se encuentra en el rango de las 21:00 y 06:00

# especificando valor de hora pero no minutos
# hour = df['fecha_hora'].dt.hour
# cond_noche = (hour >= 21) | (hour <= 6)
# df.loc[cond_noche, 'rayos_uv'] = 'BAJO'

# especificando un rango horario con horas exactas
# se extrae y almacena los valores de hora correspondientes a la columna fecha_hora
time = df['fecha_hora'].dt.time

from datetime import time as dtime
# valores de hora que determinan el rango horario y su condicion correspondiente
start, end = dtime(21, 0, 0), dtime(6, 0, 0)
cond_noche = df['fecha_hora'].dt.time >= start
cond_early = df['fecha_hora'].dt.time <= end

# variable que almacena el rango horario
horario_nocturno = cond_noche | cond_early

# se localiza los registros dentro del rango horario y se asigna el nuevo valor en la columna rayos_uv
df.loc[horario_nocturno, 'rayos_uv'] = 'NULO'
df.head(15)
```

```python
df['rayos_uv'].value_counts()
```

Si se realiza un nuevo vistazo rapido a los datos, se puede observar que siguen existiendo registros incongruentes entre columnas. Por ejemplo, en la medicion con id 8e576280 se registro un indice de rayos uv ALTO, lo cual es irracional teniendo en cuenta que el sensor realizo ese registro a las 06:18 de la mañana. Por consiguiente, no es logico que poco despues del amanecer se registre nivel de rayos uv ALTO. Para esta normalizacion de los datos, se establecera un nivel de radiacion uv BAJO en todas aquellas mediciones relizadas en el horario de 06:01 a 08:00.

```python
# especificar los horarios que determinan el rango horario y su condicion correspondiente
start, end = dtime(6, 1, 0), dtime(8, 0, 0)
cond_first_hour = df['fecha_hora'].dt.time >= start
cond_early_morning = df['fecha_hora'].dt.time <= end

# variable que almacena el rango horario
horario_mañana = cond_first_hour & cond_early_morning

# localizar los registros dentro del rango horario y asignar el nuevo valor en la columna rayos_uv
df.loc[horario_mañana, 'rayos_uv'] = 'BAJO'
df.head(15)
```

Dado que las mediciones fueron realizadas durante el verano, se puede determinar que el horario de mayor radiacion uv es entre las 11:00 y las 18:00 aproximadamente. Por consiguiente, fuera de ese horario y antes de las 21:00 el nivel de rayos uv debe ser MODERADO o BAJO. Consultar la cantidad de veces que se registro cada nivel de rayos uv en el rango horario de 18:00 a 21:00.

```python
start, end = dtime(18, 0, 0), dtime(21, 0, 0)
mask = df['fecha_hora'].dt.time.between(start, end)
atardecer = df[mask]
atardecer
```

De las mediciones realizadas en el rango horario especificado, tres de las mismas tienen asociado un nivel de rayos uv ALTO. Modificar esos tres registros y asignar nivel de rayos uv MODERADO.

```python
start, end = dtime(18, 0, 0), dtime(21, 0, 0)
value_rayos_uv = 'ALTO'
value_to_assign = 'MODERADO'
mask_time = df['fecha_hora'].dt.time.between(start, end)
mask_rayos_uv = df['rayos_uv'] == value_rayos_uv
df.loc[mask_time & mask_rayos_uv, 'rayos_uv'] = value_to_assign
```
